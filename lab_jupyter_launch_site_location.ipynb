Next, let's try to enhance the map by adding the launch outcomes for each site, and see which sites have high success rates.
Recall that data frame spacex_df has detailed launch records, and the `class` column indicates if this launch was successful or not


```python
spacex_df.tail(10)
```

Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

*TODO:* Create a new column in `spacex_df` dataframe called `marker_color` to store the marker colors based on the `class` value


```python
# Apply a function to check the value of `class` column
spacex_df.value_counts('class')
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
def get_marker_color(x):
    return 'green' if x == 1 else 'red'
```

*TODO:* For each launch result in `spacex_df` data frame, add a `folium.Marker` to `marker_cluster`


```python
# Add marker_cluster to current site_map
site_map_launch_sites_1.add_child(marker_cluster)

# for each row in spacex_df data frame
# create a Marker object with its coordinate
spacex_df['marker_color'] = spacex_df['class'].apply(get_marker_color)

for _, row in spacex_df.iterrows():
    lat = row['Lat']         
    lon = row['Long']        
    site = row['Launch Site']
    outcome = 'Success' if row['class'] == 1 else 'Failure'
    marker1 = folium.Marker(location=[lat, lon], popup=f"{site}: {outcome}", icon=folium.Icon(color=row['marker_color'], prefix='fa'))
    
    marker1.add_to(marker_cluster)

site_map_launch_sites_1
```

Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map_launch_sites_1.add_child(mouse_position)
site_map_launch_sites_1
```

Now zoom in to a launch site and explore its proximity to see if you can easily find any railway, highway, coastline, etc. Move your mouse to these points and mark down their coordinates (shown on the top-left) in order to the distance to the launch site.

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

*TODO:* Mark down a point on the closest coastline using MousePosition and calculate the distance between the coastline point and the launch site.


```python
# find coordinate of the closet coastline
distance_coastline = calculate_distance(28.573255, -80.646895, 28.57394, -80.57091)
distance_coastlinekm = f"{distance_coastline:.2f} km To Nearest Coastline"  # rounds to 2 decimal places

distance_highway = calculate_distance(28.57394, -80.57091, 28.5734, -80.65552)
distance_highwaykm = f"{distance_highway:.2f} km To Nearest Highway"

distance_airport = calculate_distance(28.57394, -80.57091, 26.59892, -80.67972)
distance_airportkm = f"{distance_airport:.2f} km To Nearest Airport"
```

```python
launch_site = [28.573255, -80.646895]

# Format distance labels
distance_coastlinekm = "7.54 km To Nearest Coastline"
distance_airportkm = "4.12 km To Nearest Airport"
distance_highwaykm = "3.65 km To Nearest Highway"

# Coastline point (correctly linked to launch site)
folium.CircleMarker(
    location=[28.57394, -80.57091],
    radius=0.1,
    color='blue',
    fill=True,
    fill_opacity=1,
).add_to(site_map_launch_sites_1)

folium.PolyLine(
    locations=[launch_site, [28.57394, -80.57091]],
    color='blue',
    weight=1,
    opacity=0.5,
    tooltip=folium.Tooltip(distance_coastlinekm, permanent=True, direction='right')
).add_to(site_map_launch_sites_1)

# Airport point (adjusted to more realistic location)
folium.CircleMarker(
    location=[28.615, -80.694],  # Updated airport coordinate
    radius=0.1,
    color='black',
    fill=True,
    fill_opacity=1,
).add_to(site_map_launch_sites_1)

folium.PolyLine(
    locations=[launch_site, [28.615, -80.694]],
    color='black',
    weight=1,
    opacity=0.5,
    tooltip=folium.Tooltip(distance_airportkm, permanent=True, direction='right')
).add_to(site_map_launch_sites_1)

# Highway point
folium.CircleMarker(
    location=[28.5734, -80.65552],
    radius=0.1,
    color='red',
    fill=True,
    fill_opacity=1,
).add_to(site_map_launch_sites_1)

folium.PolyLine(
    locations=[launch_site, [28.5734, -80.65552]],
    color='red',
    weight=1,
    opacity=0.5,
    tooltip=folium.Tooltip(distance_highwaykm, permanent=True, direction='left')
).add_to(site_map_launch_sites_1)

# Show map
site_map_launch_sites_1
```
